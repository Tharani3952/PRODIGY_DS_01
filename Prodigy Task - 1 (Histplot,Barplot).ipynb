```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('E:/Titanic.csv')
df.head()
```

```python
# Histplot of Gender
sns.histplot(df['Sex'],kde = True, color='blue')
plt.title('Distribution of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python





#Barplot of Gender
g = df['Sex'].value_counts()
plt.figure(figsize=(8, 6))
sns.barplot(x=g.index, y=g.values, palette="Accent")
plt.title('Distribution of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Histplot of Age 
sns.histplot(df['Age'],kde = True,color = "brown")
plt.title("Distribution of Age")
plt.xlabel=("Age")
plt.ylabel("Count")
plt.show()
```

```python
# Barplot of Age
h = df['Age'].value_counts()
sns.barplot(x = h.index ,y=h.values,palette="Accent")
plt.title('Distribution of Age')
plt.ylabel('Frequency')
plt.show()
```
